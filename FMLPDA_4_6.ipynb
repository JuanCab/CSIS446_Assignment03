### Step 1: Get the Code and Data You Will Need

The following cell will download the ID3 code and the data for this problem.  Run the cell below.

```python
# Import needed packages and functions
from pathlib import Path
import requests

# Download the necessary files
url_root = "https://raw.githubusercontent.com/JuanCab/CSIS446_Assignment03/refs/heads/main/data/"
# list of files to download
files_to_download = ["ID3.py", "cancer_assessment.csv"]

for file_name in files_to_download:
    url = url_root + file_name
    local_filename = file_name

    # Check if the file already exists to avoid re-downloading
    if not Path(local_filename).is_file():
        print(f"Downloading {file_name} from {url}...")
        # Download the file using requests library
        r = requests.get(url)
        # Check if the request was successful, if not, raise an error
        r.raise_for_status()

        # Save the content to a local file
        with open(local_filename, "wb") as f:
            f.write(r.content)
            print(f"File '{local_filename}' downloaded.")
```
